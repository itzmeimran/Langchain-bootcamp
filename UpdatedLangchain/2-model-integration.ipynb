# Streaming And Batch

### Streaming

Most models can stream their output content while it is beign generated. By displaying output progressively, stream significantly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yeilds output chunks as they are produced. You can use a loop to process each chunk in real-time

```python

for chunk in gemini.stream('What is RAG'):
    print(chunk.content, end='', flush=True)
```

## Batch

Batching is a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel

```python
response = gemini.batch([
    "What is RAG",
    "What is Langchain",
    "What is Vector Database"
])

for res in response:
    print(res.content)
    print("-" * 50)
```
